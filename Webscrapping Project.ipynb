# WEBSCRAPPING FROM A WEBSITE

```python
#importing relevant libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup 
```

```python
#fetching webpage

url = 'https://projects.fivethirtyeight.com/soccer-predictions/global-club-rankings/'
response = requests.get(url)
content = response.content
content
```

```python
#creating BeautifulSoup object

soup = BeautifulSoup(content, 'html.parser')
```

```python
#Extracting club names

clubs = soup.find_all('td', class_ = 'team')

names = [club.find('div', 'name').text
        for club in clubs]
names
```

```python
#extracting league names

leagues = soup.find_all('td', 'league drop-5')
league = [league.find('a', 'name').text
         for league in leagues]
league
```

```python
#extracting club's offensive power

offense = soup.find_all('td', 'num rating offense drop-6')

off = [off.find('div').text
      for off in offense]

off
```

```python
#extracting club's defensive power

defense = soup.find_all('td', 'num rating defense drop-6')
dff = [dff.find('div').text
      for dff in defense]
dff
```

```python
#extracting club's position in ranks.
ranks = soup.find('tbody')
rank = [rank.find('td', 'num ranking',text_attribute=False).text
          for rank in ranks]
rank
```

```python
#Create a dictionary with the extracted data

clubranking = {'Position': rank, 
               'Club_Name':names,
              'Leagues' : league,
              'Offensive_Power' : off,
              'Defensive_Power' : dff}

clubranking
```

```python
#creating pandas dataframe using the dictionary above

Clubs = pd.DataFrame(clubranking)
Clubs
```

```python
#Getting the list of Unique leagues

unique_leagues = Clubs['Leagues'].unique()
unique_leagues
```

```python
#Getting the number of teams represented per league 

num_teams_per_league = Clubs.groupby('Leagues')['Club_Name'].count().sort_values().reset_index(name='number')
num_teams_per_league
```

```python
#Visualing the individual leagues and the number of teams from it in a Horizontal BarChart

plt.figure(figsize= [10,8])
plt.barh(num_teams_per_league['Leagues'], num_teams_per_league['number'])
plt.xlabel('Number of Teams', fontsize = '14')
plt.ylabel('INDIVIDUAL LEAGUES', fontsize = '16')
plt.title('LEAGUES BY NUMBER OF TEAMS', fontsize ='26')
plt.savefig('Leagues BY teams.png')
```

From the above Bar chart fig 1, we can see the number of teams that each league captured in the dataset has. The number of teams were sorted in descending order therefore giving us the inference that 'Super League' produced the highest number of teams totalling 41 teams whereas the 'UEFA Champions Leagues' gave just 3 teams sitting at the bottom of the list. 

```python
#Getting the team with the highest attacking prowress

Attack_high = Clubs.groupby('Club_Name')['Offensive_Power'].max().sort_values(ascending = False).reset_index(name = 'High Offensive Power')
Attack_high.head(15)
```

```python
#Getting the team with the Lowest attacking prowress

Attack_low = Clubs.groupby('Club_Name')['Offensive_Power'].min().sort_values().reset_index(name = 'Low Offensive Power')
Attack_low.head(15)
```

```python
#Getting team with the highest defensive power

Defense_high = Clubs.groupby('Club_Name')['Defensive_Power'].max().sort_values(ascending=False).reset_index(name = 'High Defensve Power')
Defense_high.head(15)
```

```python
#Getting the team with lower defensive power

Defense_low = Clubs.groupby('Club_Name')['Defensive_Power'].min().sort_values().reset_index(name = 'Low Defensive Power')
Defense_low.head(15)
```

```python
#Getting the average defensive power per league

avg_def_power = Clubs.groupby('Leagues')['Defensive_Power'].mean().sort_values().reset_index(name = 'Avg Defence Power')
avg_def_power
```

```python
#visualizing the average defensive power of each league in a horizontal bar chart

plt.figure(figsize = [10, 8])
plt.barh(avg_def_power['Leagues'], avg_def_power['Avg Defence Power'], color= 'green')
plt.xlabel('AVG Defence Power', fontsize= '14')
plt.ylabel('INDIVIDUAL LEAGUES', fontsize = '16')
plt.title('LEAGUES AND RESPECTIVE DEFENCE POWER', fontsize= '26')
plt.savefig('Leagues vs Defence.png')
```

looking at the chart above, the quickest deduction from it is that on the average that the league with the highest average defence style of play is the 'A-league' with an average defence rating of 2.167. On the other hand, the 'La-Liga' comes lowest in defensive football play pattern with just a score of 0.85 

```python
#Getting the average offensive power per league

avg_off_power = Clubs.groupby('Leagues')['Offensive_Power'].mean().sort_values().reset_index(name = 'Avg Offensive Power')
avg_off_power
```

```python
#visualizing the average defensive power of each league in a horizontal bar chart

plt.figure(figsize = [10, 8])
plt.barh(avg_off_power['Leagues'], avg_off_power['Avg Offensive Power'], color= 'red')
plt.xlabel('AVG Offensive Power', fontsize= '14')
plt.ylabel('INDIVIDUAL LEAGUES', fontsize = '16')

plt.title('LEAGUES AND RESPECTIVE OFFENSIVE POWER', fontsize= '26')
plt.savefig('league vs offensive.png')
```

This red colored chart indicates the ratings of the various leagues as regards their attacking prowress. It can be deduced that the league with the most attacking gameplay style is the 'La-Liga' with 1.825 rating while the 'League Two' sits as the least attacking play style league with just 0.379.

```python
#Creating a new dataframe from the avg defence and avg offence by merging the both ON leagues

Combo = pd.merge(avg_def_power, avg_off_power, on='Leagues', how='inner').sort_values(by= [], ascending= False)

Combo
```

```python
#Creating a grouped bar chart that combines both the defensive and offensive prowresses

offence = Combo['Avg Offensive Power']
defence = Combo['Avg Defence Power']
y = np.arange(32)

plt.figure(figsize =[10,14])
width = 0.4
plt.barh(y-0.2, offence, width, color = 'red', label = 'offence')
plt.barh(y+0.2, defence, width, color = 'green', label = 'defence')
plt.yticks(y, Combo['Leagues'])

legend1 = plt.legend(loc= 'lower right')

# Create custom handles and labels for the second legend
line1, = plt.plot([], [], linestyle='-', color='red')
line2, = plt.plot([], [], linestyle='-', color='green')
handles = [line1, line2]
labels = ['offence', 'defence']

plt.xlabel('Scores', fontsize = '14')
plt.ylabel('INDIVIDUAL LEAGUES', fontsize = '16')
plt.title('LEAGUES BY their Offensive and Defensive powers', fontsize = '20')
plt.savefig('Leagues BY Combo.png')
```

Taking a look at the grouped bar chart, we can see a side by side comparison of each leagues defensive and offensive prowress. At a glance, one can decode what style of play is predominant in a league where it be a defensive or and offensive. 

```python
#Creating a scatterplot showing the correlation between the defensive power and offensive power of the various teams

plt.scatter( Combo['Avg Defence Power'], Combo['Avg Offensive Power'])
plt.xlabel('Avg Defence Power')
plt.ylabel('Avg Offensive Power')
plt.title('Correlation between Defensive and Offensive Prowress of Leagues')

correlation = np.corrcoef(Combo['Avg Defence Power'], Combo['Avg Offensive Power'])[0, 1]
plt.text(0.6, 0.9, f"Correlation: {correlation:.2f}", transform=plt.gca().transAxes)
plt.savefig('CorrelationPlot.png')
```

Looking at the scatterplot depicting the Correlation between the Avg Offensive Power and the Avg Defensive Power of each league, one will see that the correlation coefficient given is -0.80. This indicates a Strong Negative Coeefficient. What this number indicates is that as the Offensive Play style rating of a league increases, their defensive rating decreases and vice versa.
